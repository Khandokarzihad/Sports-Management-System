```python
# Zihad's Section



# View Primary Team Players

def primary_team_list():
     print("\n===== Primary Team =====\n")
     print("\nSl.no   Name                    Age    Preferred Role      Nationality\n")
     for i in range(len(primary_team)):
       print(" {0:6} {1:23} {2:6} {3:19} {4:15}".format(str(i + 1), primary_team[i][0],str(primary_team[i][1]),primary_team[i][2],primary_team[i][3]))

def view_players_primary():
  primary_team_list()
  view_player_exit = input("\nEnter any input to go back: ")



# View Secondary Team Players


def secondary_team_list():
    print("\n===== Secondary Team =====\n")
    print("\nSl.no   Name                    Age    Preferred Role      Nationality\n")
    for i in range(len(secondary_team)):
       print(" {0:6} {1:23} {2:6} {3:19} {4:15}".format(str(i+1), secondary_team[i][0],str(secondary_team[i][1]),secondary_team[i][2],secondary_team[i][3]))

def view_players_secondary():
  secondary_team_list()
  view_player_exit = input("\nEnter any input to go back: ")





# Add Players to the Primary Team


def add_players_primary():
  while True:
    print("\n===== Add Players in Primary Team =====\n")
    print("How many Players do you want to add?\n")
    print("Enter 0 to go back")
    player_add_num = int(input("Enter: "))
    if player_add_num == 0:
      break
    for i in range(player_add_num):
      name = input("Name: ")
      age = input("Age: ")
      preferred_role = input("Preferred Role: ")
      nationality = input("Nationality: ")
      player_details = [name, age, preferred_role, nationality]
      primary_team.append(player_details)
      print(f"{name}'s details were added\n")




# Add Players to the Secondary Team


def add_players_secondary():
  while True:
    print("\n===== Add Players in Secondary Team =====\n")
    print("How many Players do you want to add?\n")
    print("Enter 0 to go back")
    player_add_num = int(input("Enter: "))
    if player_add_num == 0:
      break
    for i in range(player_add_num):
      name = input("\nName: ")
      age = input("Age: ")
      preferred_role = input("Preferred Role: ")
      nationality = input("Nationality: ")
      player_details = [name, age, preferred_role, nationality]
      secondary_team.append(player_details)
      print(f"{name}'s details were added\n")





# Delete Primary Team Players


def del_players_primary():
  while True:
    print("\n===== Remove Players from Primary Team =====\n")
    print("How many Players do you want to remove?\n")
    print("Enter 0 to go back")
    player_del_num = int(input("Enter: "))
    if player_del_num == 0:
      break
    primary_team_list()
    for i in range(player_del_num):
      print("\nWhich player do you want to remove?\n")
      player_del_index = int(input("Enter: "))
      del primary_team[player_del_index - 1]
      primary_team_list()



# Delete Secondary Team Players


def del_players_secondary():
  while True:
    print("\n===== Remove Players from secondary Team =====\n")
    print("How many Players do you want to remove?\n")
    print("Enter 0 to go back")
    player_del_num = int(input("Enter: "))
    if player_del_num == 0:
      break
    secondary_team_list()
    for i in range(player_del_num):
      print("\nWhich player do you want to remove?\n")
      player_del_index = int(input("Enter: "))
      del secondary_team[player_del_index - 1]
      secondary_team_list()






# Team Menu


def team_menu(x, y):
  while True:
    print("\n===== Team Menu =====\n")
    print("1. Primary Team")
    print("2. Secondary Team")
    print("0. Back")
    choice_team = input("Enter: ")
    if choice_team == "1":
      x()
    elif choice_team == "2":
      y()
    elif choice_team == "0":
      break
    else:
      print("Invalid choice. Please try again.")




# Team Management menu

def team_management ():
  while True:
      print("\n===== Team Management =====\n")
      print("1. View Players")
      print("2. Add Players")
      print("3. Remove Players")
      print("0. Back")

      choice_f = input("Enter: ")

      if choice_f == "1":
        team_menu(view_players_primary, view_players_secondary)
      elif choice_f == "2":
        team_menu(add_players_primary, add_players_secondary)
      elif choice_f == "3":
        team_menu(del_players_primary, del_players_secondary)
      elif choice_f == "0":
        main_menu_admin()
      else:
        print("Invalid choice. Please try again.")











# Admin Log In


def admin_login():
  while True:
    admin_pass = "A1B2C3abc"
    print("\n===== Admin Login =====\n")
    password = input("Password: ")

    if password == admin_pass:
      print("Login successful!")
      main_menu_admin()
      break
    else:
      print("Incorrect password. Please try again.")



# User Log In


def user_login():
    while True:
       user_pass = "a1b2c3ABC"
       print("\n===== User Login =====\n")
       password = input("Password: ")

       if password == user_pass:
         print("Login successful!")
         main_menu_user()
         break
       else:
         print("Incorrect password. Please try again.")



# Authentication


def authenticate_user():
    while True:

      print("\n===== Log In =====\n")
      print("Choose your role:\n")
      print("1. Admin")
      print("2. User")
      print("\nEnter 0 to exit")

      choice = input("Enter your choice: ")

      if choice == "1":
        admin_login()
      elif choice == "2":
        user_login()
      elif choice == "0":
        print("\nExiting the system. Goodbye!")
        main()
      else:
        print("Invalid choice. Please try again.")







# Main Function


def main():
  while True:
    print("\n . . . . . . . . . . . . . . . . . . . . . . . . . . .")
    print(" . . . . . . . . . . . . . . . . . . . . . . . . . . .")
    print(" . . . . . . . . . . . . . . . . . . . . . . . . . . .")
    print(" . . . . . . . . . . . . . . . . . . . . . . . . . . .")
    print(" . . . . . . SPORTS TEAM MANAGEMENT SYSTEM . . . . . .")
    print(" . . . . . . . . . . . . . . . . . . . . . . . . . . .")
    print(" . . . . . . . . Enter 0 to continue . . . . . . . . .")
    print(" . . . . . . . . . . . . . . . . . . . . . . . . . . .")
    print(" . . . . . . . . . . . . . . . . . . . . . . . . . . .")
    print(" . . . . . . . . . . . . . . . . . . . . . . . . . . .")
    print(" . . . . . . . . . . . . . . . . . . . . . . . . . . .\n")

    enter_choice = input()
    if enter_choice == "0":
      authenticate_user()
```
